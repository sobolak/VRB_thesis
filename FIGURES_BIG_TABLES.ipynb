```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Problem': ['CVRP', 'VRPSDP', 'VRPB', 'VRPTW'],
    'Road distance': [48.345, 48.165, 55.755, 38.830],
    'Geographical distance': [23.551, 23.182, 27.070, 19.193],
}

ci = {
    'Road distance': [4.931, 5.317 , 5.378 , 4.533 ],
    'Geographical distance': [1.322, 1.322 , 1.384, 1.235],
}

df = pd.DataFrame(data)
df_ci = pd.DataFrame(ci)

fig, ax = plt.subplots(figsize=(14, 6))
plt.grid()

df_plot = df.set_index('Problem')
df_ci_plot = df_ci.set_index(df_plot.index)
df_plot.plot(kind='bar', ax=ax, yerr=df_ci_plot.T.values, capsize=4)

ax.set_ylabel('Cost function values')

plt.legend(title='', loc='upper right', bbox_to_anchor=(0.2, 1))


plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    'Number of points': [10,15,20,30],
    'Road distance': [49.255, 63.459, 70.160, 82.352],
    'Geographical distance': [23.370, 27.861, 31.074, 36.312],
    'Time cost matrix': [112.850, 145.450, 165.100, 196.650],
}

ci = {
    'Road distance': [5.502, 4.047, 2.554, 3.110],
    'Geographical distance': [1.486, 1.456, 1.193, 1.450],
    'Time cost matrix':[10.578, 8.112, 6.139, 7.066],
}

df = pd.DataFrame(data)
df_ci = pd.DataFrame(ci)

# Categories for the first plot
categories1 = ['Road distance', 'Geographical distance']
fig1, ax1 = plt.subplots(figsize=(10, 6))
width = 0.2
x = np.arange(len(categories1))

# Plot for Road distance and Geographical distance
for i, points in enumerate(df['Number of points']):
    ax1.bar(x + i * width, df.iloc[i, 1:3], width, label=f'{points} points', yerr=df_ci.iloc[i, :2], capsize=4)

ax1.set_ylabel('Cost function values')
ax1.set_xticks(x + width)
ax1.set_xticklabels(categories1)
ax1.legend(title='Number of points')
ax1.set_title('Road and Geographical Distance')

# Plot for Time cost matrix
fig2, ax2 = plt.subplots()
categories2 = ['Time cost matrix']
x = np.arange(len(categories2))

for i, points in enumerate(df['Number of points']):
    ax2.bar(x + i * width, df.iloc[i, 3:], width, label=f'{points} points', yerr=df_ci.iloc[i, 2:], capsize=4)

ax2.set_ylabel('Cost function values')
ax2.set_xticks(x + width)
ax2.set_xticklabels(categories2)
ax2.legend(title='Number of points')
ax2.set_title('Time Cost Matrix')

plt.show()
```
